```python
# Valores nulos

nulos = df.isnull().sum()
tabela_nulos = pd.DataFrame({
    "coluna": nulos.index,
    "qtd_nulos": nulos.values,
    "percentual (%)": (nulos.values / len(df)) * 100
})
print(tabela_nulos)
```

```python
# Livros com 0 páginas

livros_zero_pag = df[df["paginas"] == 0]
quant = livros_zero_pag.shape[0]
print("Quantidade de livros sem páginas: ", quant)
```

```python
# Quantos livros distintos foram publicados em cada ano cronologicamente

livros_ano = df["ano"].value_counts()
livros_ano = livros_ano.sort_index()
print(livros_ano)
```

```python
# Crie uma coluna faixa_paginas com as categorias: Curto, Médio e Longo

df["faixa_paginas"] = df["paginas"].apply(
    lambda n: "Curto" if n < 150 else ("Médio" if n <= 350 else "Longo")
)
df[["paginas", "faixa_paginas"]].head()
```

```python
# Remova todos os registros onde paginas = 0 e salve em df_limpo.

sujo = df.shape[0]
df_limpo = df[df["paginas"] > 0].copy()
depois = df_limpo.shape[0]
removido = sujo - depois

print("Quantidade de registros removidos: ", removido)
```

```python
# Preencha os nulos da coluna ano com a mediana e converta para inteiro.

mediana_ano = df_limpo["ano"].median()
df_limpo["ano"] = df_limpo["ano"].fillna(mediana_ano).astype(int)
```

```python
# Crie uma coluna decada representando a década de publicação

df_limpo["decada"] = (df_limpo["ano"] // 10) * 10
df_limpo[["ano", "decada"]].head()
```

```python
# Média de páginas por década

media_pags_decada = df_limpo.groupby("decada")["paginas"].mean().sort_index()
print(media_pags_decada)
```
